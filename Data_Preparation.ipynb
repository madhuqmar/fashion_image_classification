## Data Preparation

```python
import vogue
import os
import cv2
import numpy as np
```

```python
vogue.designer_show_to_download_images('miu-miu', 'spring-2024-ready-to-wear', './images')
```

```python
vogue.designer_show_to_download_images('helmut-lang', 'fall-2024-ready-to-wear', './images')
```

```python
labels = ['miu-miu', 'helmut-lang']
img_size = 224
```

```python
data_dir = "./images"
image_path = os.path.join(data_dir, labels[0], 'spring-2024-ready-to-wear')
print(image_path)
```

```python
class_num = labels.index(labels[0])
print(class_num)
```

```python
data = [] 
for img in os.listdir(image_path):
    try:
        img_arr = cv2.imread(os.path.join(image_path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)
```

```python
image_array = np.array(data)
print(image_array[0][0][0])
```
